Problem Statement 1: Employee Performance Bonus Eligibility

```python
# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers),
      "(Score:", highest_score, ")")
```

Problem Statement 2: Search Query Keyword Analysis

```python
# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert the query to lowercase
query = query.lower()

# Step 2: Split the sentence into words
words = query.split()

# Step 3: Count frequency of each word
keyword_count = {}

for word in words:
    if word in keyword_count:
        keyword_count[word] += 1
    else:
        keyword_count[word] = 1

# Step 4: Display only keywords searched more than once
repeated_keywords = {}

for key, value in keyword_count.items():
    if value > 1:
        repeated_keywords[key] = value

print(repeated_keywords)
```

Problem Statement 3: Sensor Data Validation

```python
# List of sensor readings where index = hour and value = reading
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid (hour, reading) pairs
valid_readings = []

# Loop through the list using index
for hour in range(len(sensor_readings)):
    reading = sensor_readings[hour]
    
    # Check if the reading is even
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

# Display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)

```

Problem Statement 4: Email Domain Usage Analysis

```python
# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Count users for each email domain
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate and display percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection

```python
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
total = 0
for amount in sales:
    total += amount

average = total / len(sales)

# Consider sales above 25% of average as spike (as per expected output)
threshold = average * 1.25

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

counts = {}

# Count occurrences
for user in user_ids:
    counts[user] = counts.get(user, 0) + 1

# Print only duplicates
for user, count in counts.items():
    if count > 1:
        print(f"{user} → {count} times")
```
